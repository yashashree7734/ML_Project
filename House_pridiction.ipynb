##  BQ1: To identify the topmost city which can help to gain optimum profit.

```python
class Business_questions1:
    def optimum_profit_city(self):
      c1,c2,c3,c4,c5           = 0,0,0,0,0
      mum,deli,kol,bang,hydr   = [],[],[],[],[]
      sum1,sum2,sum3,sum4,sum5 = 0,0,0,0,0

      # To count the individual city and getting its index value
      for i in range(len(city)):
        if city[i] == "Mumbai":
          c1 = c1+1
          mum.append(i)
        elif city[i] =="Delhi":
          c2 = c2+1
          deli.append(i)
        elif city[i] =="Kolkata":
          c3 = c3+1
          kol.append(i)
        elif city[i] == "Bangalore":
          c4 = c4+1
          bang.append(i)
        else:
          c5=c5+1
          hydr.append(i)

      # Suming up the number of houses available for each citie
      total_houses = {"Mumbai":sum(mum),"Delhi":sum(deli),"Kolkata":sum(kol),"Hydrabad":sum(hydr),"Bangalore":sum(bang)}

      # Totaling the rent value for each city
      for i in mum:
        sum1 = sum1+rent[i]
      for i in deli:
        sum2 = sum2+rent[i]
      for i in kol:
        sum3 = sum3+rent[i]
      for i in hydr:
        sum4 = sum4+rent[i]
      for i in bang:
        sum5 = sum5+rent[i]

      # Storing the value in dictionary
      optimum_profit = {"Mumbai":sum1,"Delhi":sum2,"Kolkata":sum3,"Hydrabad":sum4,"Bangalore":sum5}

      print("Total number of houses for rent in each city")
      for k,v in total_houses.items():
        print(k, " : " ,v)

      print()

      print("Total number of profit obtain from each City")
      for k,v in optimum_profit.items():
        print(k," : ",v)

    def create_graph(self):
      print()
      plt.bar(city,rent,color = "#87ace8")
      plt.xlabel("Citys")
      plt.ylabel("Rent")
      plt.title("Rent distribution based on City")
      plt.show()
```

```python
q1 = Business_questions1()
q1.optimum_profit_city()
q1.create_graph()
```

According to the above mentioned information, Mumbai, Delhi and Kolkata are the cities that can yield the highest profit, which is approximately 128700 rupees. Since there are around 258 apartments in Mumbai that are accessible for rental purposes, we can first look to rent our homes in Mumbai before moving on to Delhi and Kolkata.

### BQ4: To estimate the period of time a family or group of people spends in a particular property.

```python
def contract_broke():

  #importing required libararies
  import statistics
  from statistics import median_grouped

  least_time = []
  ind        = []
  tenat      = []

  # Calculating the median grouped to collect values below the median in data sets
  std = int(median_grouped(time))

  # Colleting time based on the median value
  for i in time:
    if i < std:
      least_time.append(i)
      ind.append(time.index(i))

  # Grouping tenant according to the time
  for i in ind:
    tenat.append(tenant[i])

  print("Precetange of Family likely to broke contarct ",[((tenat.count("Family")/len(tenant)))*100])
  print("Precetange of Bachleros likely to broke contarct ",[(tenat.count("Bachelors")/len(tenant))*100])

contract_broke()
```

```python
sn.countplot(x =time,hue=tenant )
plt.show()
```

From the about output it is noticible that, Bachelors are more likely to sign a contract with a term between 6,8 and 23. However, most of the family would rather not remain in the falt for an extended period of time. Therefore, families are more prone than bachelors to break the agreement and move out. As a result, having a family in a falt can be advantageous because property values rise about every 6-7 months.
